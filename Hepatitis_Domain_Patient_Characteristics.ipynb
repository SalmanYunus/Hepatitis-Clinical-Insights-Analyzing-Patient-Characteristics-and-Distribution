```python
for column in df.columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Drop rows with NaN values
df.dropna(inplace=True)


# 6. The code identifies and ranks the top 10 most frequent symptoms based on their occurrence in the data. 
#    This information is crucial for understanding which symptoms are most commonly experienced by patients with Hepatitis for early diagnosis & priortizing patient care.
#    The code generates a bar chart that visually represents the frequency of each symptom


# Symptom names
symptoms = ["FATIGUE", "MALAISE", "ANOREXIA", "LIVER_BIG", "LIVER_FIRM", "SPLEEN_PALPABLE" , "SPIDERS","ASCITES","VARICES"]

# Empty list for store symptom counts
symptom_counts = []

# Loop through each symptom
for symptom in symptoms:
    # Calculate the number of instances where the symptom is present (value = 1)
    symptom_count = df[df[symptom] == 1].shape[0]
    
    # Append the count and symptom name to the list
    symptom_counts.append({"symptom": symptom, "count": symptom_count})

# Sort the list by count in descending order
symptom_counts = sorted(symptom_counts, key=lambda x: x["count"], reverse=True)

# Print top 10 most frequent symptoms
print("Top 10 most frequent symptoms:")
for i in range(min(10, len(symptom_counts))):
    print(f"{i+1}. Symptom: {symptom_counts[i]['symptom']}, Count: {symptom_counts[i]['count']}")

# Create a bar chart to visualize the symptom frequency
plt.figure(figsize=(10, 6))
plt.bar([symptom["symptom"] for symptom in symptom_counts], [symptom["count"] for symptom in symptom_counts])
plt.xlabel("Symptom")
plt.ylabel("Count")
plt.title("Frequency of Symptoms")
plt.xticks(rotation=45, ha="right")
plt.show()


# 7. This code and its output delve deeper into the Hepatitis data by analyzing and visualizing the prevalence of symptoms across different age groups.
#    Analyzing Age-Specific Symptom Prevalence for Identifying age groups at risk and Understanding disease progression.
#    Below Visualization offers valuable insights into the age-specific prevalence of various symptoms in Hepatitis patients. This information can inform targeted interventions, 
#    improve patient care, and contribute to our understanding of the disease progression across different demographics.


# Initialize empty dictionary to store symptom prevalence by age group
symptom_prevalence_by_age = {symptom: {} for symptom in symptoms}

# Define age groups
age_groups = df["AGE"].unique()

# Loop through each symptom and age group
for symptom in symptoms:
    for age_group in age_groups:
        # Calculate the prevalence of the symptom in the current age group
        symptom_count_in_age_group = df[(df[symptom] == 1) & (df["AGE"] == age_group)].shape[0]
        total_count_in_age_group = df[df["AGE"] == age_group].shape[0]
        
        # Calculate and store the prevalence as a percentage
        symptom_prevalence_by_age[symptom][age_group] = (symptom_count_in_age_group / total_count_in_age_group) * 100

# Print symptom prevalence by age group
print("Symptom Prevalence by Age Group:")
for symptom, prevalence_data in symptom_prevalence_by_age.items():
    print(f"\nSymptom: {symptom}")
    for age_group, prevalence in prevalence_data.items():
        print(f"\tAge Group: {age_group}, Prevalence: {prevalence:.2f}%")
        
plt.figure(figsize=(10, 6))

for i, symptom in enumerate(symptoms):
    plt.plot(age_groups, [symptom_prevalence_by_age[symptom][age_group] for age_group in age_groups], label=symptom, color=plt.cm.Paired(i))
plt.xlabel("Age Group")
plt.ylabel("Symptom Prevalence (%)")
plt.title("Symptom Prevalence by Age Group")
plt.legend()
plt.show()

# Create a heatmap : Offers a visual summary of the data, highlighting areas with high or low prevalence for each symptom-age group combination
plt.figure(figsize=(10, 6))
data = [[symptom_prevalence_by_age[symptom][age_group] for age_group in age_groups] for symptom in symptoms]
plt.imshow(data, cmap="YlOrRd")
plt.colorbar()
plt.xticks(range(len(age_groups)), age_groups)
plt.yticks(range(len(symptoms)), symptoms)
plt.title("Symptom Prevalence by Age Group (Heatmap)")
plt.xticks(rotation=90, ha="right")
plt.show()



# 8.  This code analyzes and visualize the prevalence and occurrence of symptoms in the Hepatitis data based on gender.

symptom_prevalence_by_gender = {symptom: {} for symptom in symptoms}
symptom_counts_by_gender = {symptom: {} for symptom in symptoms}

genders = df["SEX"].unique()

for symptom in symptoms:
    for gender in genders:
        # Calculate the prevalence of the symptom in the current gender group
        symptom_count_in_gender = df[(df[symptom] == 1) & (df["SEX"] == gender)].shape[0]
        total_count_in_gender = df[df["SEX"] == gender].shape[0]

        # Calculate and store the prevalence as a percentage
        symptom_prevalence_by_gender[symptom][gender] = (symptom_count_in_gender / total_count_in_gender) * 100

# Print symptom prevalence by gender
print("\n Symptom Prevalence by Gender:")
for symptom, prevalence_data in symptom_prevalence_by_gender.items():
    print(f"\nSymptom: {symptom}")
    for gender, prevalence in prevalence_data.items():
        print(f"\tGender: {gender}, Prevalence: {prevalence:.2f}%")

# Create a bar chart
plt.figure(figsize=(10, 6))
for i, symptom in enumerate(symptoms):
    # Count the occurrences of each symptom by gender
    counts_by_gender = df[df[symptom] == 1]["SEX"].value_counts()
    # Store the counts in the dictionary for later use
    symptom_counts_by_gender[symptom] = counts_by_gender.to_dict()

    # Plot bars with different colors for each gender
    colors = ['skyblue' if gender == 1 else 'pink' for gender in counts_by_gender.index]
    plt.bar(
        counts_by_gender.index,
        counts_by_gender.values,
        label=symptom,
        color=colors,
    )


plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Symptom Counts by Gender")
plt.legend()
plt.show()
```

***`10. Analyzing Symptom Prevalence , Co-occurrences and Understanding symptom patterns.`***

```python

# 10. The code calculates the prevalence of each symptom (percentage of patients experiencing the symptom) and the co-occurrence of any two symptoms (percentage of patients experiencing both symptoms).
#     Understanding symptom patterns , Analysis and Visualization Techniques.

symptoms = [
    "FATIGUE",
    "MALAISE",
    "ANOREXIA",
    "LIVER_BIG",
    "LIVER_FIRM",
    "SPLEEN_PALPABLE",
    "SPIDERS",
    "ASCITES",
    "VARICES",
]

# Dictionaries for symptom prevalence and co-occurrence
symptom_prevalence = defaultdict(int)
symptom_co_occurrence = defaultdict(lambda: defaultdict(int))


# Calculate symptom prevalence and co-occurrence
for index, row in df.iterrows():
    for symptom in symptoms:
        try:
            symptom_present = int(row[symptom])  # Convert symptom value to integer
        except ValueError:
#             print(f"Error converting symptom value for {symptom} in row {index}.")
            continue

        symptom_prevalence[symptom] += int(symptom_present)

        for symptom2 in symptoms:
            if symptom != symptom2:
                try:
                    symptom2_present = int(row[symptom2])
                except ValueError:
#                     print(f"Error converting symptom value for {symptom2} in row {index}.")
                    continue

                if symptom2_present:
                    symptom_co_occurrence[symptom][symptom2] += int(symptom_present * symptom2_present)

# Calculate co-occurrence percentages
total_patients = len(df)
for symptom1, co_occurrence_data in symptom_co_occurrence.items():
    for symptom2, count in co_occurrence_data.items():
        symptom_co_occurrence[symptom1][symptom2] = (count / total_patients) * 100

# Analyze symptom prevalence
print("\nSymptom Prevalence:")
for symptom, count in symptom_prevalence.items():
    prevalence_percentage = (count / total_patients) * 100
    print(f"\t- {symptom}: {prevalence_percentage:.2f}%")

# Visualize co-occurrence heatmap
plt.figure(figsize=(10, 6))
data = [[value for value in symptom_data.values()] for symptom_data in symptom_co_occurrence.values()]
plt.imshow(data, cmap="YlOrRd")
plt.colorbar()
plt.xticks(range(len(symptoms)), symptoms)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.yticks(range(len(symptoms)), symptoms)
plt.title("Symptom Co-occurrence Heatmap")
plt.show()
```

***`11. Exploring Relationships between Lab Tests and Categories.`***

```python

# 11. This code investigates the distribution of different laboratory tests across various categorical variables in the Hepatitis data. It uses a combination of data cleaning, statistical analysis, and visualization techniques to provide 
#     valuable insights. The code analyzes how lab test values (BILIRUBIN, ALK_PHOSPHATE, SGOT, ALBUMIN) vary across different categorical variables like Class (Alive/Dead), gender, histology (liver inflammation), medications, and 
#     various symptoms


# Define laboratory test names
lab_tests = [
    "BILIRUBIN",
    "ALK_PHOSPHATE",
    "SGOT",
    "ALBUMIN"
]


# Define categorical variables
categorical_vars = ["Class","SEX","HISTOLOGY", "STEROID","ANTIVIRALS","FATIGUE","MALAISE","LIVER_BIG","LIVER_FIRM","SPLEEN_PALPABLE","SPIDERS","ASCITES","VARICES"]

# Read data from CSV file
df = pd.read_csv(file_path)

for column in df.columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Drop rows with NaN values
df.dropna(inplace=True)


# Create a function to plot boxplots and violin plots
def plot_distributions(data, cat_var, lab_test, figsize=(10, 6)):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=figsize)

    sns.boxplot(x=cat_var, y=lab_test, data=data, showmeans=True, ax=axes[0])
    axes[0].set_title(f"Boxplot - {lab_test} vs. {cat_var}")

    sns.violinplot(x=cat_var, y=lab_test, data=data, ax=axes[1])
    axes[1].set_title(f"Violin Plot - {lab_test} vs. {cat_var}")

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Analyze and plot distributions for each lab test across categories
for lab_test in lab_tests:
    for cat_var in categorical_vars:
        plot_distributions(df, cat_var, lab_test)
```

***`14. Identify patient subgroups with high-risk profiles for targeted interventions and preventative measures.`***

```python
# 14. This code and its output develop and analyze a patient risk score based on specific factors identified from the Hepatitis data. It then uses the score to identify high-risk patients and analyze their characteristics and outcomes.
#     The code defines a risk score based on four identified risk factors: "LIVER_BIG", "SPLEEN_PALPABLE", "BILIRUBIN", and "SGOT".
#     The code visualizes the distribution of risk scores across the patient population using a histogram and Identifying High-Risk Patients.


# Define risk score components based on findings from analysis 
risk_factors = ["LIVER_BIG", "SPLEEN_PALPABLE", "BILIRUBIN", "SGOT"]

# Calculate risk score for each patient
df["Risk Score"] = 0
for factor in risk_factors:
    # Convert to integer if necessary and handle missing values
    factor_values = pd.to_numeric(df[factor], errors='coerce')
    factor_values = factor_values.fillna(0)  # Replace missing values with 0 (adjust as needed)
    df["Risk Score"] += factor_values * 2
    
    
# Analyze distribution of risk scores
plt.figure(figsize=(8, 6))
plt.hist(df["Risk Score"], bins=10, edgecolor="black", alpha=0.7)
plt.xlabel("Risk Score")
plt.ylabel("Frequency")
plt.title("Distribution of Patient Risk Scores")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Identify high-risk patients based on score thresholds
high_risk_threshold = 6  # Adjust threshold based on analysis
high_risk_patients = df[df["Risk Score"] >= high_risk_threshold]

# Analyze characteristics and outcomes of high-risk patients
print("High-Risk Patient Demographics:")
print(high_risk_patients.groupby(["SEX", "AGE"]).size())
print("\nHigh-Risk Patient Outcomes:")
print(high_risk_patients["Class"].value_counts())

```

***`15. Evaluate the effectiveness of different treatment combinations on patient outcomes within specific subgroups.`***

```python
# 15. The code allows for a systematic evaluation of the impact of different treatment combinations on patient outcomes, considering specific subgroups defined by age and gender. 
#     The use of stratified K-fold cross-validation ensures robust evaluation, and logistic regression is used as the predictive model.

# Define outcome variable (adjust class mapping)
y = df["Class"].map({"DIE": 0, "LIVE": 1})

# Define feature variables
features = ["STEROID", "ANTIVIRALS", "AGE", "SEX"] + list(df.columns[6:15])

# Define subgroups to analyze
subgroups = [("AGE", "<40", ">40"), ("SEX", "male", "female")]

# Perform analysis for each subgroup
for group_by, group1, group2 in subgroups:
    # Filter data for specific subgroup
    df_group = df[df[group_by] == group1]

    # Define treatment combinations
    treatments = [
        ("no", "no"), ("no", "yes"), ("yes", "no"), ("yes", "yes")
    ]

    # Drop rows with missing values in features
    df_group = df_group.dropna(subset=features, how="any")

    # Perform analysis for each treatment combination
    for treatment1, treatment2 in treatments:
        # Check if samples available for current combination
        if len(df_group[(df_group['STEROID'] == treatment1) & (df_group['ANTIVIRALS'] == treatment2)]) > 0:
            # Perform stratified K-fold cross-validation
            accuracies = []
            aucs = []
            kf = StratifiedKFold(n_splits=5)
            for train_index, test_index in kf.split(df_group[features], y):
                # Train model on training data
                model = LogisticRegression()  # Change model as needed
                model.fit(df_group.loc[train_index, features], y.loc[train_index])

                # Predict outcome on test data
                predictions = model.predict(df_group.loc[test_index, features])

                # Calculate accuracy and AUC
                accuracy = accuracy_score(y.loc[test_index], predictions)
                auc = roc_auc_score(y.loc[test_index], predictions)

                accuracies.append(accuracy)
                aucs.append(auc)

            # Print results
            print(f"\t{group_by}: {group1}, {treatment1} Steroid, {treatment2} Antiviral:")
            print(f"\t\tAverage Accuracy: {np.mean(accuracies):.4f}")
            print(f"\t\tAverage AUC: {np.mean(aucs):.4f}")
        else:
            print(f"\t{group_by}: {group1}, {treatment1} Steroid, {treatment2} Antiviral: No samples available.")

    print("-" * 50)

```
